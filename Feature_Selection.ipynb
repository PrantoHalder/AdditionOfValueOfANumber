```python
#import necesssary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as splt
import seaborn
```

```python
#load the dataset
df=pd.read_csv('Mobile_Adiction.csv')#,sep=';')
df
```

```python
#check if there is any null value available
df.isna().sum()
```

```python
#fit the value into x
x=df.iloc[:,:-1]
x
```

```python
#fit the target value
y=df.iloc[:,20]
y
```

```python
# select K best algorithom for feature selection
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
FIT_FEATURES = SelectKBest(score_func=f_classif)
FIT_FEATURES.fit(x,y)
```

```python
pd.DataFrame(FIT_FEATURES.scores_)
```
