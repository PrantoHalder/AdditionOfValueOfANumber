```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as splt
import seaborn
```

```python
df=pd.read_csv('Mobile_Adiction.csv')
df
```

```python
df['Are you addicted with phone'].value_counts()
```

```python
df['Gender'].value_counts()
```

```python
df['Age'].value_counts()
```

```python
df=df.drop(['Gender'],axis=1)
df=df.drop(['Age'],axis=1)
df=df.drop(['Education'],axis=1)
df=df.drop(['Occupation'],axis=1)
df=df.drop(['Live with family'],axis=1)
df=df.drop(['Motive about Using mobile'],axis=1)
df=df.drop(['smoker'],axis=1)
df=df.drop(['Friends influence you to use phone'],axis=1)
df=df.drop(['Time you prefer for using mobile'],axis=1)

df
```

```python
df.corr()
```

```python
splt.figure(figsize=(16,10))

splt.xticks(fontsize=14)
splt.yticks(fontsize=14)
seaborn.heatmap(df.corr(),annot=True)
splt.show()
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,11]
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.3,random_state=1)
xtrain
```

```python
from sklearn.ensemble import RandomForestClassifier
Rclf=RandomForestClassifier()
Rclf.fit(xtrain,ytrain)
Rclf.score(xtest,ytest)
```

```python
from sklearn import tree
Clf = tree.DecisionTreeClassifier()
Clf.fit(xtrain,ytrain)
Clf.score(xtest,ytest)
```

```python
from sklearn.ensemble import AdaBoostClassifier
clf=AdaBoostClassifier()
clf.fit(xtrain,ytrain)
clf.score(xtest,ytest)
```

```python
from sklearn.svm import SVC
model=SVC(gamma='auto')
model.fit(xtrain,ytrain)
model.score(xtest,ytest)
```
